```python
from gensim.models import Word2Vec
```

```python
x = filter
y = data['sentiment']
```

```python
X=[]
for i in x:
    X.append(word_tokenize(i))
```

```python
bog=Word2Vec(X, min_count=1,vector_size=100, window=5)
```

```python
def vec(token):
    vector=[bog.wv[word] for word in token if word in bog.wv]
    return np.mean(vector,axis=0) if vector else np.zeros(bog.vector_size)
```

```python
x=[vec(token) for token in x]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print('Classification Report:\n', classification_report(y_test, y_pred))
```

```python
#skipgram
```

```python
sg=Word2Vec(X, min_count=1,vector_size=100, window=5,sg=1)
```

```python
def sg_vec(token):
    vector=[sg.wv[word] for word in token if word in sg.wv]
    return np.mean(vector,axis=0) if vector else np.zeros(sg.vector_size)
```

```python
x=[sg_vec(token) for token in x]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 42)
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print('Classification Report:\n', classification_report(y_test, y_pred))
```
